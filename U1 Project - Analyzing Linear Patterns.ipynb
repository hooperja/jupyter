### **Task #5: Modify the code and generate the pattern** 

The code below uses a `for` loop. Read through the code to see how the for-loop works.

* Read the code below. (Yes - actually "read" it first!)
* Modify the code so that the plant starts at 2 inches and grows 0.5 inches each day, to match this situation above.

````
days = list(range(8)) # Create a list of days
heights = [5] # Create a list with the starting value as the only entry
growth = 1.5

# Notes:
# In the for-loop below, we do not need to append a height for the day0
# days[1:] is the same as the days list, without day0.
# heights[-1] is always the most recent height.
for day in days[1:]:
  heights.append(heights[-1]+growth)

from tabulate import tabulate
table = [
         ["Days"]+days,
         ["Height (inches)"]+heights,
]

print(tabulate(table))
````

### **Task #10: Label the plot** 
The code below creates a plot of the plant growth. Your job is to make the plot "well-labeled".

* The x-axis should be labeled "Days"
* The y-axis should be labeled "Plant Height (inches)"
* The title should be "Plant Height vs Days"

````
days = list(range(21))
heights = [2]
growth = 0.5

for day in days[1:]:
  heights.append(heights[-1]+growth)

import matplotlib.pyplot as plt
plt.scatter(days, heights)
plt.xlabel("x")
plt.ylabel("y")
plt.title("y vs x")
plt.grid(True)
plt.show()
````
